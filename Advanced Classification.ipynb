## - Feature Engineering

Since the variables EducationField and JobRole has several classes, we will try to further explore these varaibles to see their distribution by Attrition

```python
#exploring EducationField variable
EducationFieldCrosstab = pd.crosstab(data['EducationField'], data['Attrition'], margins=True)
EducationFieldCrosstab
```

```python
EducationFieldCrosstab.div(EducationFieldCrosstab["All"], axis=0)
```

It can be seen that Human Resources and Technical degree, Life Sciences and Medical have high similarities in their distributions and hence we will combine them

```python
# Combining job roles with high similarities together
data['EducationField'].replace(['Human Resources','Technical Degree'],value= 'HR-TD',inplace = True)
data['EducationField'].replace(['Life Sciences','Medical'],value= 'LS-M',inplace = True)
```

```python
print(data['EducationField'].value_counts())
```

We will do the same for JobRole variable as shown in the code sections below

```python
# Combine classes in JobRole variable by seeing their distribution
JobRoleCrossTab = pd.crosstab(data['JobRole'], data['Attrition'], margins=True)
JobRoleCrossTab
```

```python
JobRoleCrossTab.div(JobRoleCrossTab["All"], axis=0)
```

```python
# Combining job roles with high similarities together
data['JobRole'].replace(['Human Resources','Laboratory Technician'],value= 'HR-LT',inplace = True)
data['JobRole'].replace(['Research Scientist','Sales Executive'],value= 'RS-SE',inplace = True)
data['JobRole'].replace(['Healthcare Representative','Manufacturing Director'],value= 'HE-MD',inplace = True)
```

```python
print(data['JobRole'].value_counts())
```

```python
#Encoding Attrition variable from Yes/No to 1/0
data['Attrition'].replace(["Yes","No"],[1,0],inplace=True)
data.head(3)
```

### - Splitting the data into training and test datasets

```python
#creating dataframes from target variable and features
X = final_data.drop(['Attrition'], axis = 1)
y = final_data['Attrition']
```

```python
#splitting the data into training set and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state = 0, stratify=y)
```

* About Stratify : Stratify parameter makes a split so that the proportion of values in the sample produced will be the same as the proportion of values provided to parameter stratify.
* For example, if variable y is a binary categorical variable with values 0 and 1 and there are 25% of zeros and 75% of ones, stratify=y will make sure that your random split has 25% of 0's and 75% of 1's
* Let's now confirm that the class distribution of the target variable is uniform across both the test and training sets - 


```python
# Checking distribtution of Target varaible in training set
y_train.value_counts()[1]/(y_train.value_counts()[0]+y_train.value_counts()[1])*100
```

```python
y_test.value_counts()[1]/(y_test.value_counts()[0]+y_test.value_counts()[1])*100
```

Hence we can confirm that the stratifcation of the data was successful

## - Model Development

We will now move on to the model development phase where we build several machine learning models to predict the exit status of the employee and also compare the results of the different models. 
For the purpose of this project we will be building and evaluating the following models

1. Logistic Regression
2. Naive Bayes
3. Decision Tree Classifier
3. Random Forest Classifier

#### 1. Logistic Regression

```python
#MODEL TRAINING

#importing library for logistic regression
from sklearn.linear_model import LogisticRegression

#create a logistic regression classifier
logreg = LogisticRegression()

#training the model on the training set
logreg.fit(X_train, y_train)
```

```python
# Prediction on test data
y_pred = logreg.predict(X_test)
```

```python
#calculating the accuracy, precision and the recall
acc_logreg = round(metrics.accuracy_score(y_test, y_pred)*100, 2)
pre_logreg = round(metrics.precision_score(y_test, y_pred)*100,2)
rec_logreg = round(metrics.recall_score(y_test, y_pred)*100, 2)
print('Total Accuracy: ', acc_logreg)
print('Precision: ', pre_logreg)
print('Recall: ', rec_logreg)
```

```python
#creating confusion matrix
from sklearn.metrics import auc,confusion_matrix
def draw_cm(actual, predicted):
    cm = confusion_matrix( actual, predicted, [1,0]).T
    sns.heatmap(cm, annot=True,  fmt='.2f', xticklabels = ["Yes","No"] , yticklabels = ["Yes","No"] )
    plt.ylabel('Predicted')
    plt.xlabel('Actual')
    plt.show()
```

```python
draw_cm(y_test, y_pred)
```

#### 2. Naive Bayes

```python
# Import library for Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB

# Create a Gaussian Classifier
model = GaussianNB()

# Train the model using the training sets 
model.fit(X_train,y_train)
```

```python
# Prediction on test set
y_pred = model.predict(X_test)
```

```python
# Calculating the accuracy, precision and the recall
acc_nb = round( metrics.accuracy_score(y_test, y_pred) * 100, 2 )
pre_nb = round( metrics.precision_score(y_test, y_pred) * 100, 2 )
rec_nb = round( metrics.recall_score(y_test, y_pred) * 100, 2 )
print( 'Total Accuracy : ', acc_nb)
print( 'Precision : ',pre_nb)
print( 'Recall : ',rec_nb)
```

```python
# Confusion matrix 
draw_cm(y_test, y_pred)
```

#### 3. Decision Tree Classifier

```python
# Import Decision tree classifier
from sklearn.tree import DecisionTreeClassifier

# Create a Decision tree classifier model
clf = DecisionTreeClassifier()

# Train the model using the training sets 
clf.fit(X_train, y_train)
```

```python
# Plotting feature importance
feature_importance = pd.DataFrame([X_train.columns, clf.tree_.compute_feature_importances()])
feature_importance = feature_importance.T.sort_values(by = 1, ascending=False)[1:10]
sns.barplot(x=feature_importance[1], y=feature_importance[0])
# Add labels to the graph
plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.title("Visualizing Important Features")
plt.legend()
plt.show()
```

```python
# Prediction on test set
y_pred = clf.predict(X_test)
```

```python
# Calculating the accuracy, precision and the recall
acc_dt = round( metrics.accuracy_score(y_test, y_pred) * 100, 2 )
pre_dt = round( metrics.precision_score(y_test, y_pred) * 100, 2 )
rec_dt = round( metrics.recall_score(y_test, y_pred) * 100, 2 )
print( 'Total Accuracy : ', acc_dt )
print( 'Precision : ',pre_dt  )
print( 'Recall : ',rec_dt  )
```

```python
# Confusion matrix
draw_cm(y_test, y_pred)
```

#### 4. Random Forest Classifier

```python
# Import library of RandomForestClassifier model
from sklearn.ensemble import RandomForestClassifier

# Create a Random Forest Classifier
rf = RandomForestClassifier()

# Train the model using the training sets 
rf.fit(X_train,y_train)
```

```python
# Plotting feature importance
feature_imp = pd.Series(rf.feature_importances_,index= X_train.columns).sort_values(ascending=False)
feature_imp=feature_imp[0:10,]
sns.barplot(x=feature_imp, y=feature_imp.index)
# Add labels to the graph
plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.title("Visualizing Important Features")
plt.legend()
plt.show()
```

```python
# Prediction on test data
y_pred = rf.predict(X_test)
```

```python
# Calculating the accuracy, precision and the recall
acc_rf = round( metrics.accuracy_score(y_test, y_pred) * 100 , 2 )
pre_rf = round( metrics.precision_score(y_test, y_pred) * 100 , 2 )
rec_rf = round( metrics.recall_score(y_test, y_pred) * 100, 2 )
print( 'Total Accuracy : ', acc_rf )
print( 'Precision : ',pre_rf )
print( 'Recall : ',  rec_rf)
```

```python
# Confusion metrix
draw_cm(y_test, y_pred)
```

## - Model Comparison/Interpreting the results


Having built the various machine learning models, it is now time to see how the stack up against each other

```python
print('Logistic Regression: ',acc_logreg)
print('Naive Bayes: ' ,acc_nb)
print('Decision Tree Classifier: ' ,acc_dt)
print('Random Forest Classifier: ' ,acc_rf)
```
